```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales=pd.read_csv("C:/Users/NANDAKISHORE BOINI/OneDrive/Desktop/sales_data_with_discounts.csv")
sales.head()
```

```python
numeric_columns = ['Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

plt.figure(figsize=(12, 6))

for col in numeric_columns:
    plt.plot(sales[col], label=col)

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Sales Data Columns')
plt.legend()
plt.show()
```

```python
numeric_columns = ['Volume','Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

for col in numeric_columns:
    plt.figure(figsize=(10, 4))
    plt.plot(sales[col])
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title(f'{col}')
    plt.show()
```

```python
from scipy.stats import zscore
# List of numeric columns to apply z-score
numeric_columns = ['Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

# Apply z-score normalization
sales[numeric_columns] = sales[numeric_columns].apply(lambda x: zscore(x))

# Display the first few rows to verify
print(sales.head())
```

```python
numeric_columns = ['Volume','Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
for col in numeric_columns:
    sales[col] = zscore(sales[col])

# Display the first few rows to verify
print(sales.head())
```

```python
numeric_columns = ['Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
for col in numeric_columns:
    sales[col] = zscore(sales[col])

# Plotting z-score values
plt.figure(figsize=(12, 6))

# Plot each numeric column
for col in numeric_columns:
    plt.plot(sales[col], label=col)

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Z-Score')
plt.title('Z-Score of Various Sales Data Columns')
plt.legend()
plt.show()
```

```python
for col in numeric_columns:
    sales[col] = zscore(sales[col])
numeric_columns = ['Volume','Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
# Plotting each z-score column separately
for col in numeric_columns:
    plt.figure(figsize=(10, 4))
    plt.plot(sales[col])
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.title(f'Z-Score of {col}')
    plt.show()
```
